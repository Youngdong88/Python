### **스카이스포츠 15/16 시즌 url 접속 (토트넘)**

```python
from selenium import webdriver
import sys

wd = webdriver.Chrome('chromedriver.exe')
wd.get("https://www.skysports.com/tottenham-hotspur-results/2015-16")
```

### **접속 페이지에서 'div' 태그만 긁어오기**

```python
div_tag = wd.find_elements_by_tag_name('div')
div_tag
```

### **'div'태그 하위에 있는 'a' 태그 / 'h5' 태그 긁어오기**

```python
a_tag = []
h5_tag = []

for i in div_tag:
    a = i.find_elements_by_tag_name('h5')
    b = i.find_elements_by_tag_name('a')
    a_tag.append(b)
    h5_tag.append(a)

# 이중리스트 이기 때문에 합쳐줘야 함
h5_tag = sum(h5_tag, [])
a_tag = sum(a_tag, [])
```

---

### **h5_tag에서 프리미어 리그만 뽑아오기**

```python
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse

web_url = 'https://www.skysports.com/tottenham-hotspur-results/2015-16'

epl_html = []

with urllib.request.urlopen(web_url) as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')    

    a = soup.find_all('div', 'fixres__body')

    epl_html.append(a)
    
epl_html = sum(epl_html, [])
epl_html = " ".join(map(str, epl_html))
```

### **'epl_links' 정규화**

```python
import re

epl_links = re.findall("Premier League.*\s.*\s.*\s", epl_html)

for i in range(0, len(epl_links)):
    epl_links[i] = epl_links[i].replace('"', "-")
    epl_links[i] = epl_links[i][157:-3]
    
epl_links
```

## **선수명 불러오기 (디코딩 포함)**

```python
wd = webdriver.Chrome('chromedriver.exe')
wd.get("https://www.premierleague.com/players?se=42&cl=21")
```

```python
player_list = wd.find_elements_by_class_name('playerName')

player_name_0 = []

for i in range(0, len(player_list)):
    name = player_list[i].text
    player_name_0.append(name)

player_name_0
```

### **선수 이름에 있는 라틴어 디코딩**

```python
from unicodedata import normalize, category

player_name = []

for i in player_name_0:
    a = ''.join([c for c in normalize('NFD', i) if category(c) != 'Mn'])
    player_name.append(a)
    
player_name
```

---

### **데이터 테이블 만들기**

```python
import pandas as pd
import numpy as np

data = {'Game_No': np.nan, 'link': epl_links}

df_1516 = pd.DataFrame(data)

df_1516
```

```python
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse

date_html = []

for i in df_1516['link']:
    with urllib.request.urlopen(i) as response:
        html = response.read()
        soup = BeautifulSoup(html, 'html.parser')    

        a = soup.find_all('li', 'match-header__detail-item')

        date_html.append(a)

# 이중 리스트 제거
date_html = sum(date_html, [])

# 리스트를 문자열로
date_html = " ".join(map(str, date_html))

#  pm으로 시작하고, 첫번째 엔터(\n)까지 찾는다
date = re.findall("pm.*\s", date_html)

# 'pm "삭제하고, 처음 등장하는 "<" 인덱스 까지 자른다
for i in range(0, len(date)):
    date[i] = date[i][3:]
    date[i] = date[i][:date[i].find("<")]

date
```

### **게임 날짜 및 선수명 헤더로 넣어주기**

```python
df_1516['Game_No'] = date

for i in player_name:
    df_1516[i] = np.nan
    
df_1516
```
